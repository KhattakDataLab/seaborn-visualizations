```python
# import all the library 
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# create a dictionary 

dic={

'age'  :  [21,22,25,27,29,28],
'Salry':  [50000,55000,60000,80000,90000,100000],
'Dept' :  ['Finance','Marketing','IT','HR','IT','HR']
    
}

df=pd.DataFrame(dic)
```

```python
df
```

# Scatter Plot 

#### Scatter plot can be used to show the relation ship b/w continious variable

```python
sns.scatterplot(x='age',y='Salry',data=df)
plt.title('Age & Salary ')
plt.show()
```

# Bar Plot

```python
# bar plot show the average salary by dept

sns.barplot(x='Dept', y='Salry', data=df)
plt.title('Average Salary by Department')
plt.show()
```

# Box plot 

#### box plot show us the distribution of data 

```python
sns.boxplot(x='Dept', y='Salry', data=df)
plt.title('Salary Distribution by Department')
plt.show()
```

# Customizing Seaborn Plots

```python
 #Adding a hue for department
sns.scatterplot(x='age', y='Salry', hue='Dept', data=df)
plt.title('Age vs Salary by Department')
plt.show()
```

# Heatmap

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data
data = {
    'age': [25, 32, 47, 51, 62],
    'salary': [50000, 60000, 80000, 120000, 150000],
    'department': ['HR', 'IT', 'Finance', 'IT', 'Finance']
}

# Create DataFrame
df = pd.DataFrame(data)

# Compute correlation matrix (only for numeric columns)
corr = df[['age', 'salary']].corr()

# Plot heatmap
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
